# Reducing multicollinearity

```python
df.corr()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.corr(), annot=True)
```

Graph is a bit tight but I need to do some feature engineering here. Possibly to much multicollinearity. I will be using logistic regression since this is a classification problem and I don't want multicollinearity to get in the way. 

```python
#Making a smaller df with target variable and features that have a direct impact on that.
df2 = df[['blueWins','blueWardsPlaced', 'blueWardsDestroyed', 
                'blueFirstBlood', 'blueKills', 'blueDeaths', 'blueAssists', 
                'blueDragons', 'blueHeralds', 'blueTowersDestroyed', 
                'blueTotalMinionsKilled', 'blueTotalJungleMinionsKilled',
                'redWardsPlaced', 'redWardsDestroyed', 
                'redFirstBlood', 'redKills', 'redDeaths', 'redAssists', 
                'redDragons', 'redHeralds', 'redTowersDestroyed', 
                'redTotalMinionsKilled', 'redTotalJungleMinionsKilled']]
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df2.corr(), annot=True)
```

This graph still has multicollineraity for example redFirstBlood, redKills, and redDeaths are all direct inverses of blueFirstBlood, blueKills, and blueDeaths. I will drop these features for logistic regression.

```python
df2.drop (columns = ['redKills','redDeaths','redFirstBlood'], inplace = True)
```

```python
df2.columns
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df2.corr(), annot=True)
```

I feel much happier with this correlation matrix. Blue assist and blue kills have high correlation but I will leave it as they are heavily tied. Same with red assists and blue deaths. I will leave these features because they do have heavy impact on a matches outcome.

# Machine Learning

#### Logistic Regression

```python
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y = df2['blueWins']
```

```python
y
```

```python
X = df2.drop('blueWins', axis = 1, inplace = False)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)

print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred=logistic.predict(X_test)
confmat = confusion_matrix(y_true=y_test, y_pred=y_pred)
confusion = pd.DataFrame(confmat, index=['Predicted Class Positive', 'Predicted Class Negative'],
                         columns=['True Class Postivie', 'True Class Negative'])

confusion
```

```python
print (classification_report(y_test, y_pred, digits=3))
```

```python
## Accuracy
## How often is the classifier correct?
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print ("Accuracy score: %.3f" %(acc*100))
```

```python
## Precision
## Ability of the classifier to avoid labeling a class as a member of another class
from sklearn.metrics import precision_score

pcs = precision_score(y_test, y_pred)
print ("Precision: %.3f" %(pcs*100))
```

```python
## Recall
## Recall the ability of the classifier to correctly identify the current class
from sklearn.metrics import recall_score

rcs = recall_score(y_test, y_pred)
print ("Recall: %.3f" % (rcs*100))
```

I figured the accuracy would be rather low. This game relys on a lot of factors and of course come from behind wins are likely. Taking just a snapshot of the match at 10 minutes in a match that could last 30-40 minutes doesn't show everything. Champion selection also matters and that information isn't in this dataset. Lastly, a minion named baron that spawns at 20 minutes into the game isn't here and getting baron first is extremely important.
